```python
import matplotlib.pyplot as plt
days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
steps=[4000,6000,5500,7000,8000,10000,9000]
plt.plot(days,steps,marker='o',ms=8)
plt.title("Weekly Step Count")
plt.xlabel("Days")
plt.ylabel("Steps")
plt.grid()
max_steps=max(steps)
max_day=days[steps.index(max_steps)]
plt.text(max_day,max_steps,f"Peak:{max_steps}")
plt.show()
```

```python
import matplotlib.pyplot as plt
days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
data_usage=[2,3.5,3,4,5,6.5,6]
plt.plot(days,data_usage,ls='--',lw=2,color='hotpink')
plt.title("Weekly Interest Data Usage")
plt.xlabel("Days")
plt.ylabel("Data Usage (GB)")
plt.grid()
max_val=max(data_usage)
max_day=days[data_usage.index(max_val)]
plt.text(max_day,max_val,f"Peak:{max_val} GB")
plt.show()
```
