```python
import matplotlib.pyplot as plt
days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
study_hours=[2,3,4,3,5,6,4]
sleep_hours=[8,7,6,7,5,6,7]
plt.plot(days,study_hours,marker='o',label="Study Hours")
plt.plot(days,sleep_hours,marker='s',label="Sleep Hours")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.title("Study vs Sleep Pattern")
plt.legend()
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.title("Study vs Sleep Relationship")
plt.show()
```
