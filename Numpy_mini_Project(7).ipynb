```python
import numpy as np
users=1000
prob=[0.4,0.3,0.2,0.1]
result=np.random.multinomial(users,prob)
features=["Chat","Video","Games","Settings"]
print("Feature usage counts:")
for i in range(4):
    print(features[i]+":",result[i])
max_index=np.argmax(result)
min_index=np.argmin(result)
print("\nMost used feature:",features[max_index])
print("Least used features:",features[min_index])
print("Total users:",np.sum(result))
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
downloads=500
scale=2
wait=np.random.exponential(scale=scale,size=downloads)
print("Average waiting time:",np.mean(wait))
print("Max waiting time:",np.max(wait))
print("Min waiting time:",np.min(wait))
less1=np.sum(wait<1)
between=np.sum((wait>=1)&(wait<=3))
more3=np.sum(wait>3)
print("\n<1 sec:",less1)
print("1-3 sec:",between)
print(">3 sec:",more3)
sns.histplot(wait,kde=True)
plt.title("Download Waiting time(Exponential Distribution")
plt.show()
```
