```python
import pandas as pd
data={
    "UsageHours":[2,4,6,8,10],
    "ScreenTime":[1,3,5,7,9],
    "AppsUsed":[5,10,15,20,25],
    "BatteryDrain":[10,20,35,50,65]
}
df=pd.DataFrame(data)
print("Data\n")
print(df)
corr=df.corr()
print("\nCorrelation Matrix\n")
print(corr)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Calories": [200, 250, 300, 280, 350, 400, 370]
}

df = pd.DataFrame(data)

# Set Day as index for better plotting
df.set_index("Day", inplace=True)

# Plot line graph
df.plot(kind="line", marker="o", title="Daily Calories Burned")

# Labels
plt.xlabel("Day")
plt.ylabel("Calories")

# Show graph
plt.show()
```
